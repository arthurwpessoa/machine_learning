```python
# Required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Reading pastry dataset
df = pd.read_csv('../datasets/coffee_shop/pastry.csv')

df.head()
```

```python
# Drops null values
df.dropna(inplace = True)
```

```python
# Replaces column to remove special character
df = df.drop(columns={'% waste'})
```

```python
# Adds a new column with the equivalent day of week of each date
# MONDAY: 0 - SUNDAY 6
df["transaction_weekday"] = pd.to_datetime(df['transaction_date'], format = '%m/%d/%Y').dt.dayofweek

df = df.drop(columns={'transaction_date'})
```

```python
df["quantity_consumed"] = df['start_of_day'] - df['quantity_sold']
```

```python
df.head()
#create a new column for predicted values for each product_iod and transaction_weekday
```

### Data Analysis

```python
#  Displays the data correlation between the features.
#  Strong correlation between the start_of_day and the waste
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), annot = True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
plt.show()
```

## Define test data

```python
y = df['quantity_consumed']

X = df[['product_id','transaction_weekday', 'waste']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Linear Regression

```python
lr = LinearRegression()
```

```python
# Training test data
model = lr.fit(X_train,y_train)
```

```python
# Coefficients for product_id and transaction_weekday, respectively.
coeffecients = pd.DataFrame(lr.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
# Predicting values based on test data.
predictions_test = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print(np.sqrt(mean_squared_error(y_test,predictions_test)))
print(r2_score(y_test, predictions_test))
```

```python
# Predicting values based on trained data.
predictions_train = lr.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
print(np.sqrt(mean_squared_error(y_train,predictions_train)))
print(r2_score(y_train, predictions_train))
```

```python
#Include features - underestimating
# X_test, y_test, predictions - get product_id - 69
plt.scatter(y_test, predictions_test)
plt.xlabel('Test Data')
plt.ylabel('Predicted Data')
```

```python
#Checking distribution of residuals.
plt.hist((y_test-predictions_test));
```

```python
#item id - 69, 
myvals = np.array([69, 3, 0]).reshape(1, -1)
print("Number of this pastry item needed on a Wednesday: " + str(model.predict(myvals)[0]))
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge(alpha=0.01)
rr_model = rr.fit(X_train, y_train) 
pred_train_rr= rr.predict(X_test)
```

```python
#Include features - underestimating
# X_test, y_test, predictions - get product_id - 69
rr_y,rr_p = [], []
for i_X_test, value in enumerate(X_test["product_id"]):
    if(value == 69):
        rr_y.append(y_test.values[i_X_test])
        rr_p.append(pred_train_rr[i_X_test])
        print("test: " + str(y_test.values[i_X_test]) + " prediction: " + str((pred_train_rr[i_X_test])))
plt.scatter(rr_y, rr_p)
plt.xlabel('Test Data')
plt.ylabel('Predicted Data')
```
